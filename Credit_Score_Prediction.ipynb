#### There is outliers in the dataset we need to deal with it

```python
for col in df.columns:
    if df[col].dtypes == 'int64' or df[col].dtypes == float:
        lower_fence = df[col].quantile(0.25)
        upper_fence = df[col].quantile(0.75)
        
        IQR = upper_fence - lower_fence
        
        df = df[(df[col]>= lower_fence - 1.5*IQR) & (df[col]<=upper_fence + 1.5*IQR)]
```

#### Now, we will make model and check for the accuracy like which model is giving better accuracy

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 34)
```

#### Now, starting with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### SupportVectorClassifier

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(accuracy_score(y_pred, y_test))
```

```python
from sklearn.svm import SVC
svc = SVC(probability=True)
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(accuracy_score(y_pred, y_test))
```

#### GaussianNaiveBayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
y_pred = nb.predict(x_test)
accuracy_score(y_test, y_pred)
```

#### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred)
```

#### BaggingClassifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier(base_estimator = tree, n_estimators = 100, random_state=35)
bc.fit(x_train, y_train)
y_pred = bc.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### ExtraTreeClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
et = ExtraTreesClassifier()
et.fit(x_train, y_train)
y_pred = et.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
abc.fit(x_train, y_train)
y_pred= abc.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=100, random_state=25)
gb.fit(x_train, y_train)
y_pred = gb.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### ExtremeGradientBoosting


```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

#### RidgeClassifier

```python
from sklearn.linear_model import RidgeClassifier
rc = RidgeClassifier()
rc.fit(x_train, y_train)
y_pred = rc.predict(x_test)
print(accuracy_score(y_test, y_pred))
print(rc.score(x,y))
```

```python
# predicting the probability of getting good credit Score
rf_probs = rf.predict_proba(x_test)[:,1]
et_probs = et.predict_proba(x_test)[:,1]
bc_probs = bc.predict_proba(x_test)[:,1]
#rc_probs = rc.predict_proba(x_test)[:,1] RidgeClassifier has no attribute predictproba
xgb_probs = xgb.predict_proba(x_test)[:,1]
abc_probs = abc.predict_proba(x_test)[:,1]
knn_probs = knn.predict_proba(x_test)[:,1]
nb_probs = nb.predict_proba(x_test)[:,1]
svc_probs = svc.predict_proba(x_test)[:,1]
dt_probs = dt.predict_proba(x_test)[:,1]
lr_probs = lr.predict_proba(x_test)[:,1]
gb_probs = gb.predict_proba(x_test)[:,1]
```

##### AUC ROC Score

```python
from sklearn.metrics import roc_auc_score
ras = roc_auc_score
rf_auc = ras(y_test, rf_probs)
et_auc = ras(y_test, et_probs)
bc_auc = ras(y_test, bc_probs)
xgb_auc = ras(y_test, xgb_probs)
abc_auc = ras(y_test, abc_probs)
knn_auc = ras(y_test, knn_probs)
nb_auc = ras(y_test, nb_probs)
svc_auc = ras(y_test, svc_probs)
dt_auc = ras(y_test, dt_probs)
lr_auc = ras(y_test, lr_probs)
gb_auc = ras(y_test, gb_probs)
```

```python
from sklearn.metrics import roc_curve
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, threshold = roc_curve(y_test, random_probs)
```

```python
print('Random Forest: Area under ROC = %0.2f' %rf_auc)
print('Extra Tree: Area under ROC = %0.2f' %et_auc)
print('Extra Gradient Boosting: Area under ROC = %0.2f' %xgb_auc)
print('Bagging Classifier: Area under ROC = %0.2f' %bc_auc)
print('Ada Boost Classifier: Area under ROC = %0.2f' %abc_auc)
print('K Nearest Neighbors: Area under ROC = %0.2f' %knn_auc)
print('Naive Bayes: Area under ROC = %0.2f' %nb_auc)
print('Support Vector Classifier: Area under ROC = %0.2f' %svc_auc)
print('Decision Tree: Area under ROC = %0.2f' %dt_auc)
print('Logistic Regression: Area under ROC = %0.2f' %lr_auc)
print('Gradient Boosting : Area under ROC = %0.2f' %gb_auc)
```

```python
nb_fpr,nb_tpr,thresholds=roc_curve(y_test,nb_probs)
svc_fpr,svc_tpr,thresholds=roc_curve(y_test,svc_probs)
dt_fpr,dt_tpr,thresholds=roc_curve(y_test,dt_probs)
rf_fpr,rf_tpr,thresholds=roc_curve(y_test,rf_probs)
et_fpr,et_tpr,thresholds=roc_curve(y_test,et_probs)
lr_fpr,lr_tpr,thresholds=roc_curve(y_test,lr_probs)
gb_fpr,gb_tpr,thresholds=roc_curve(y_test,gb_probs)
knn_fpr,knn_tpr,thresholds=roc_curve(y_test,knn_probs)
abc_fpr,abc_tpr,thresholds=roc_curve(y_test,abc_probs)
bc_fpr,bc_tpr,thresholds=roc_curve(y_test,bc_probs)
xgb_fpr,xgb_tpr,thresholds=roc_curve(y_test,xgb_probs)
```

```python
plt.plot(nb_fpr,nb_tpr, label="Naive Bayes (AUROC=%0.2f)"%nb_auc)
plt.plot(svc_fpr,svc_tpr, label="Support Vector Classifier (AUROC=%0.2f)"%svc_auc)
plt.plot(dt_fpr,dt_tpr, label="Decision Tree Classifier (AUROC=%0.2f)"%dt_auc)
plt.plot(rf_fpr,rf_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%rf_auc)
plt.plot(et_fpr,et_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%et_auc)
plt.plot(lr_fpr,lr_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%lr_auc)
plt.plot(gb_fpr,gb_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%gb_auc)
plt.plot(knn_fpr,knn_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%knn_auc)
plt.plot(abc_fpr,abc_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%abc_auc)
plt.plot(bc_fpr,bc_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%bc_auc)
plt.plot(xgb_fpr,xgb_tpr, label="Random Forest Classifier (AUROC=%0.2f)"%xgb_auc)

plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend()
plt.show()
```
